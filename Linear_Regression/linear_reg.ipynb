```python
import pandas as pd
import numpy as np
```

```python

data = {
    "Date": [
        "01.01.24","01.12.23","01.11.23","01.10.23","01.09.23",
        "01.08.23","01.07.23","01.06.23","01.05.23","01.04.23",
        "01.03.23","01.02.23"
    ],
    "Starting (USD)": [
        16800,15900,15800,16100,16300,
        16800,15900,15800,16150,16300,
        16200,16300
    ],
    "Ending (USD)": [
        16500,16100,15300,16200,15700,
        16400,16200,15500,16100,15800,
        16200,15700
    ]
}
df = pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()
```

```python
x=df[['Starting (USD)']]
x.head()
```

```python
y=df[['Ending (USD)']]
y.head()
```

```python
x.mean()
```

```python
y.mean()
```

```python
import matplotlib.pyplot as plt
#best fit line ta ei point(red) diye jabee
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Starting (USD)'],df['Ending (USD)'])
```

```python
from sklearn.linear_model import LinearRegression
#LinearRegression er object create
lr=LinearRegression()
```

```python
lr.fit(x,y)#model ke train korano(m,c er man calculate hobee)
```

```python
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
#y=mx+c(by equation calculate the predicted value for x=16800)
Y=m*16800+c
Y
```

```python
lr.predict([[16800]])#predict value by model
```

```python
df['predicted_y']=lr.predict(x)
df.head()
```

```python
plt.plot(x, df['predicted_y'])
#plt.scatter(x.mean(), y.mean(), color='red', s=100)
#plt.autoscale()
plt.xlabel('starting usd')
plt.ylabel('predicted ending usd')
```

```python
df['lose']=df['Ending (USD)']-df['predicted_y']
df.head()
```

```python
#mse and mae
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse=mean_squared_error(df['Ending (USD)'],df['predicted_y'])
mse
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mae=mean_absolute_error(df['Ending (USD)'],df['predicted_y'])
mae
```

```python
#by manually using equation(mae)
sum(abs(df['lose']))/len(x)
```

```python
plt.plot(x, df['predicted_y'])
plt.scatter(df['Starting (USD)'],df['Ending (USD)'])
plt.scatter(df['Starting (USD)'],df['predicted_y'])

#predicted value (brown color point hocche blue point er predicted point )
```

```python
#r2 score{accuracy for regression}
from sklearn.metrics import r2_score
y_true = df['Ending (USD)']
y_pred = df['predicted_y']
```

```python
r2_score(y_true,y_pred)
```

**2nd example(salary prediction)**

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Salary Data.csv')
df.head()
```

```python
x=df[['Experience Years']]
x.head()
```

```python
y=df[['Salary']]
y.head()
```

```python
x_mean=x.mean()
x_mean
```

```python
y_mean=y.mean()
y_mean
```

```python
import matplotlib.pyplot as plt
#ei red point diye best fit line ta jabee
plt.scatter(x_mean,y_mean,color='red')
plt.scatter(df['Experience Years'],df['Salary'])
plt.xlabel('experience year')
plt.ylabel('salary')
```

**linear regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()#creating object
```

```python
#training model(calculating m and c)
lr.fit(x,y)
```

```python
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
#calculating predicted value by equation(y=mx+c)
Y=m*1.1+c
Y
```

```python
#predicted value by model
lr.predict([[1.1]])
```

```python
df['predicted_Y']=lr.predict(x)
df.head()
```

```python
plt.plot(x,	df['predicted_Y'])
plt.scatter(x_mean,y_mean,color='green')
plt.scatter(df['Experience Years'],df['Salary'])
plt.scatter(x,df['predicted_Y'])
plt.xlabel('experience year')
plt.ylabel('salary')
#orange point are predicted value of blue ponit.
```

**loss and cost function**

```python
#loss means different of every single data point of acctual and predicted value
df['loss']=df['Salary']-df['predicted_Y']
df.head()
```

```python
#calculating loss and cost by model
#mse & mae
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(df['Salary'],df['predicted_Y'])
mse
```

```python
mae=mean_absolute_error(df['Salary'],df['predicted_Y'])
mae
```

```python
sum(abs(df['loss']))/len(x)
```
