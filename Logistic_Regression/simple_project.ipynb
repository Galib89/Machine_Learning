```python
df.drop(columns=['Ad Topic Line','City','Country','Timestamp'],inplace=True)
```

```python
import seaborn as sns
sns.boxplot(data=df,x='Clicked on Ad',y='Daily Internet Usage')
```

```python
sns.pairplot(data=df,kind='kde',hue='Clicked on Ad')
```

```python


feature=df.drop(columns=['Clicked on Ad'])
target=df['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(feature,target,train_size=.67,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
prediction=lr.predict(x_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
