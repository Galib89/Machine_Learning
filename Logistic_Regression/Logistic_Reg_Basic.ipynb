```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python


np.random.seed(0)

df = pd.DataFrame(
    data={
        'Feature': np.arange(0,101),
        'Target': np.array([0]*60 + [1]*41)
    }
)

df.head()
```

```python
df.info()
```

```python
sns.lmplot(data=df,x='Feature',y='Target')
#regression plot
#Target->Catagorical
#Feature->Linear
```

```python
df1=df.copy()
#adding some noise to visualize logistic regression
df1["Target"]=df1["Target"]+np.random.normal(0,0.03,size=len(df))
df1["Target"]=df1["Target"].clip(0,1)
#clip->ignore which value are less then 0 and greater than 1

df1.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.regplot(data=df1,x='Feature',y='Target',label='Linear regression')
sns.regplot(data=df1,x='Feature',y='Target',label='Linear regression',logistic=True,color='red')
```
