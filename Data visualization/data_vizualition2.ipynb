```python
tips=sns.load_dataset('tips')
tips.head(5)
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='time',palette='husl')
```

```python
d=tips[['size','total_bill','tip']].corr()
d
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head(5)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data=diamonds,x='carat',y='price')
plt.title("Diamond Prices by Carat")

plt.xlabel("Carat (ct)")

plt.ylabel("Price ($)")

plt.show()
```
