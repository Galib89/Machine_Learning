```python
import plotly.express as px
import seaborn as sns
```

```python
diamonds_df=sns.load_dataset('diamonds')
diamonds_df.info()
```

```python
fig=px.histogram(diamonds_df,x='price',title='Diamond plot',width=1000,height=700)
fig.show()
```

```python
diamond_mean_p=diamonds_df.groupby("cut")['price'].mean().reset_index()
diamond_mean_p
```

```python
fig=px.bar(diamond_mean_p,x='cut',y='price',color='cut')
fig.show()
```

```python
fig=px.scatter(diamonds_df,x='price',y='carat',color='cut',size='price')
fig.show()
```

```python
fig=px.box(diamonds_df,x='cut',y='price',color='cut')
fig.show()
```
